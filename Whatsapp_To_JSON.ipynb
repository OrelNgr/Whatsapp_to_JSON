```python
def Whatsapp_JsonTxt(filename):
    from datetime import datetime
    import json
    myFile=open(filename,encoding='utf8')
    lst=list()
    ids=dict()
    newid=1

    # go over the file
    try:
        for line in myFile: 
            #Edge case: take raws that only have text, and add them to the text of the last person who sent a messege
            if not line.split()[0][0].isnumeric(): 
                lst[-1]['text']=lst[-1]['text']+' '+line.strip()
                continue

            #get messeges general data 
            #find the creation raw [and skip other 'special' raws] and save general data [date,creator, and name]
            if line.find(':',17)==-1: 
                if line.find('נוצרה על ידי')!=-1:
                    creation_date=str(datetime.strptime(line.split(",")[0]+' '+line.split()[1],'%d.%m.%Y %H:%M'))
                    creator=line[line.find('נוצרה על ידי')+13:].strip('\u200f').strip('\u202a').strip().strip('\u200f').strip("\u202c")
                    chat_name=line[line.find('הקבוצה')+8:line.find('"',line.find('הקבוצה')+8)]
                continue

            #get messege specific data    
            #get the name of the messege writer
            id1=line[line.find('-')+2:line.find(':',16)]
            #give each person an id instead of his name
            ids[id1.strip("\u200f\u202a").strip("\u202c")]=ids.get(id1.strip("\u200f\u202a").strip("\u202c"),newid)
            newid+=1
            #save the text messege
            text=line[line.find(':',17)+2:].strip()
            #save the time
            datetime1=datetime.strptime(line.split(",")[0]+' '+line.split()[1],'%d.%m.%Y %H:%M')
            #append all dict data to list
            lst.append({'datetime':str(datetime1)[:-3],'id':ids[id1.strip("\u200f\u202a").strip("\u202c")],'text':text})

        #create metadata dictionary that holds general data
        metaDict={'chat_name': chat_name , 'creation_date': creation_date[:-3],
        'num_of_participants': len(ids), 'creator' : creator}

        #dictionary with the data + metadata
        allDict={'messages':lst,'metadata':metaDict}

        #save as text in JSON formatting
        try:
            nameOfFile=chat_name+'.txt'
            with open(nameOfFile, 'w',encoding='utf8') as outfile:
                json.dump(allDict, outfile, ensure_ascii=False, indent=4)
            #print a friendly messege of completion
            print()
            print('קובץ טקסט בפורמט ג'+"'"+'ייסון המכיל את פרטי קבוצת הוואטספ "'+chat_name+'" נשמר בהצלחה')
        except:
            nameOfFile='myGroupJSON.txt'
            with open(nameOfFile, 'w',encoding='utf8') as outfile:
                json.dump(allDict, outfile, ensure_ascii=False, indent=4)

            print()
            print("because of problematic name, the file name is 'myGroupJSON.txt'")
            #print a friendly messege of completion
            print('.קובץ טקסט בפורמט ג'+'ייסון המכיל את פרטי קבוצך נשמר בהצלחה')

    except:
        print("מצטער, ייצור הקובץ נכשל")
```
